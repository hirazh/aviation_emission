## Per capita aviation emissions - 2018

Aviation contribute around 2.5% of global CO2 emissions <br>
Source: [Our world in data](https://ourworldindata.org/carbon-footprint-flying)

```python
# imprts
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading data

# Domestic aviation co2 per capita 2018
domestic_aviation = pd.read_csv('data/per-capita-co2-domestic-aviation.csv')
# International aviation co2 per capita 2018
international_aviation = pd.read_csv('data/per-capita-co2-international-aviation.csv')
```

```python
domestic_aviation = domestic_aviation.rename(columns={'Per capita domestic aviation CO2': 'Per capita co2'})
domestic_aviation['Per capita co2 (kg)'] = round(domestic_aviation['Per capita co2'] * 1000, 2)
#domestic_aviation
```

```python
international_aviation = international_aviation.rename(columns={'Per capita international aviation CO2': 'Per capita co2'})
international_aviation['Per capita co2 (kg)'] = round(international_aviation['Per capita co2'] * 1000, 2)
#international_aviation
```

```python
# ploting the top 20 countries with highest emission 
# from domestic air travel
top_20_domestic = domestic_aviation.sort_values(by=['Per capita co2 (kg)'], ascending=False)[0:20].reset_index()
#top_20_domestic
```

```python
ax1, fig = plt.subplots(figsize=(14,8)) 
ax1 = sns.set_style('darkgrid')
#ax1 = sns.set_palette("Paired")
ax1 = sns.barplot(data=top_20_domestic, x='Entity', y='Per capita co2 (kg)', palette=('YlOrRd_r'))
ax1.set_xticklabels(labels=top_20_domestic['Entity'], rotation=60)
ax1.set_title('Per Capita CO2 Emission: Domestic Aviation - 2018', fontsize=18)
ax1.set_xlabel('Country', fontsize=14)
ax1.set_ylabel('Emission Per Capita (Kg)', fontsize=14)

for p in ax1.patches:
    ax1.annotate(format(p.get_height(), '.1f'), 
                 (p.get_x() + p.get_width() / 2., p.get_height()),
                ha = 'center', va = 'center',
                xytext = (0, 5),
                textcoords = 'offset points')
plt.show()
```

```python
# ploting the top 20 countries with highest emission 
# from international air travel
top_20_international = international_aviation.sort_values(by=['Per capita co2 (kg)'], ascending=False)[0:20].reset_index()
#top_20_international
```

Domestic aviation emissions are starighforwared since the flights are within the same country we can add this to country's total greenhouse gas emission. But international aviation emissions originating in one country and landing in another country its hard to hold one country responsible for the emissions.

```python
ax1, fig = plt.subplots(figsize=(14,8)) 
ax1 = sns.set_style('darkgrid')
#ax1 = sns.set_palette("Paired")
ax1 = sns.barplot(data=top_20_international, x='Entity', y='Per capita co2 (kg)', palette=('YlOrRd_r'))
ax1.set_xticklabels(labels=top_20_international['Entity'], rotation=60)
ax1.set_title('Per Capita CO2 Emission: International Aviation - 2018', fontsize=18)
ax1.set_xlabel('Country', fontsize=14)
ax1.set_ylabel('Emission Per Capita (Kg)', fontsize=14)

for p in ax1.patches:
    ax1.annotate(format(p.get_height(), '.1f'), 
                 (p.get_x() + p.get_width() / 2., p.get_height()),
                ha = 'center', va = 'center',
                xytext = (0, 5),
                textcoords = 'offset points')
plt.show()
```
